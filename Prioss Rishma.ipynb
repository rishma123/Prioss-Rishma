```python
import json
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt
import folium
import pandas as pd
```

```python
with open('loginsample.json', 'r') as json_file:
	json_load = json.load(json_file)
data = pd.DataFrame(json_load['account_activity'])
```

```python
data["lat"] = ""
data["long"] = "" 
```

```python
for index,row in data.iterrows():
    Nomi_locator = Nominatim(user_agent=row['user_agent'])
    location = Nomi_locator.geocode(row['city'])
    data.at[index,"lat"] = location.latitude
    data.at[index,"long"] = location.longitude
```

```python
m = folium.Map(location=[51.1657,10.4515], tiles="OpenStreetMap", zoom_start=6)
for index,row in data.iterrows():
    pop = "<b>City</b>:" +row['city'] + "<br>"
    pop += "<b>Date</b>:"+str(row['timestamp']) + "<br>"
    pop += "<b>Site name</b>:"+row['site_name']+ "<br>"
    pop += "<b>IP address</b> : "+row['ip_address'] + "<br>"
    pop += "<b>Action</b>: "+row['action'] + "<br>"
    pop = folium.Popup(folium.IFrame(pop),min_width=300,max_width=350)
    folium.Marker(
      location=[row['lat'], row['long']],
      popup=pop,
   ).add_to(m)
m
```
